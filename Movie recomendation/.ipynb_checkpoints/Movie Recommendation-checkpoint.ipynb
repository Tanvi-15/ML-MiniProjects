```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
# TfidfVectorizer - This is used to convert text data into numerical values
from sklearn.metrics.pairwise import cosine_similarity
import os
```

```python

# display (os.getcwd())
```

```python

movies_data =pd.read_csv('movies.csv')
# movies_data.head()
```

```python
display (movies_data.shape)
movies_data.columns
```

```python
selected_features = ['genres','keywords','tagline','cast','original_language']
print(selected_features)
```

```python
display (movies_data.info())
```

```python
display (movies_data.isna().sum())
```

```python
display (movies_data[selected_features].head())
```

```python
display (movies_data[selected_features].isna().sum())
```

```python
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
display (movies_data.head())
```

```python

combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['original_language']
display (combined_features)
```

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
display (feature_vectors.shape)
print (feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
print  (similarity)

```

```python
display(similarity.shape)
```

```python
movie_name = input(' Enter your favourite movie name : ')
```

```python

list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
len(list_of_all_titles)
```

```python
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

```python
print('Movies suggested for you : \n')
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
